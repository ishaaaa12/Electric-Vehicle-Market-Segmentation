```python
def avg_selling_price(df1, category):
    avg_price_by_category = df1.groupby(category)['selling_price'].mean().sort_values()
    plt.figure(figsize=(10, 6))
    avg_price_by_category.plot(kind='bar', color='skyblue')
    plt.title(f'Average Selling Price by {category.capitalize()}')
    plt.xlabel(category.capitalize())
    plt.ylabel('Average Selling Price')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()
```

```python
# Plot average selling price by fuel type
avg_selling_price(df1, 'fuel')
plt.savefig('avg price.png')
```

```python
# Plot average selling price by transmission type
avg_selling_price(df1, 'transmission')
plt.savefig('avg price by transmission.png')
```

# K-Means Clustering

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
features = ['selling_price', 'km_driven', 'year', 'transmission', 'seller_type', 'owner']

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), ['selling_price', 'km_driven', 'year']),
        ('cat', OneHotEncoder(), ['transmission', 'seller_type', 'owner'])
    ]
)


pipeline = Pipeline(steps=[('preprocessor', preprocessor)])


X = pipeline.fit_transform(df1[features])


sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42,n_init='auto')
    kmeans.fit(X)
    sse.append(kmeans.inertia_)


plt.plot(range(1, 11),sse, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.title('Elbow Method')
plt.savefig('elbow.jpg')
plt.show()


for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    score = silhouette_score(X, kmeans.labels_)
    print(f'Number of clusters: {k}, Silhouette Score: {score}')


k_optimal = 3
kmeans = KMeans(n_clusters=k_optimal, random_state=42)
clusters = kmeans.fit_predict(X)


df1['cluster'] = clusters


for i in range(k_optimal):
    print(f'Cluster {i}:')
    print(df1[df1['cluster'] == i].describe())
    
df1.to_csv('electric_vehicle_sales_clustered.csv', index=False)
```
